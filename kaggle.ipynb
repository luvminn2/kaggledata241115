```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVC
from sklearn.metrics import mean_squared_error, mean_absolute_error , accuracy_score, classification_report, confusion_matrix
from sklearn.ensemble import RandomForestClassifier

train_data = pd.read_csv('./Training_cleaned.csv')
```

```python
train_data
```

```python
x = train_data.drop(columns='prognosis')
y = train_data['prognosis']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

from sklearn.tree import DecisionTreeClassifier
dt_model = DecisionTreeClassifier()
dt_model.fit(x_train, y_train)

dt_y_pred = dt_model.predict(x_test)
```

```python
print("Decision Tree 예측률:", accuracy_score(y_test, dt_y_pred))
```

```python
print("Decision Tree 표:\n", classification_report(y_test, dt_y_pred))
```

```python
print("Decision Tree Confusion Matrix:\n", confusion_matrix(y_test, dt_y_pred))
```

```python
svm_model = SVC(kernel='linear')
svm_model.fit(x_train, y_train)

svm_y_pred = svm_model.predict(x_test)
```

```python
print("\nSVM 예측률:", accuracy_score(y_test, svm_y_pred))
```

```python
print("SVM Classification 표:\n", classification_report(y_test, svm_y_pred))
```

```python
print("SVM Confusion Matrix:\n", confusion_matrix(y_test, svm_y_pred))
```

```python
rf_model = RandomForestClassifier()
rf_model.fit(x_train, y_train)

rf_y_pred = rf_model.predict(x_test)
```

```python
print("\nRandom Forest 표:", accuracy_score(y_test, rf_y_pred))
```

```python
print("Random Forest Classification 표:\n", classification_report(y_test, rf_y_pred))
```

```python
print("Random Forest Confusion Matrix:\n", confusion_matrix(y_test, rf_y_pred))
```
